```python
import pymongo
import json
import requests
from bson import ObjectId

# URL de conexão ao MongoDB Atlas (substitua pela sua)
client = pymongo.MongoClient("mongodb+srv://naua61:abacaxi4512@clusterestadios.1sey2.mongodb.net/")

# Acessar o banco de dados e a coleção
db = client["games_database"]
collection = db["games"]

# Função para buscar Pokémon na PokéAPI
def fetch_pokemon_data(pokemon_id):
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_id}/'
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Erro ao buscar dados para o Pokémon {pokemon_id}")
        return None

# Exemplo de IDs de Pokémon para buscar
pokemon_ids = [1, 4, 7, 25, 151]  # Bulbasaur, Charmander, Squirtle, Pikachu, Mew

# Listar os dados dos Pokémon
pokemon_data = []
for pokemon_id in pokemon_ids:
    data = fetch_pokemon_data(pokemon_id)
    if data:
        # Organizar os dados do Pokémon
        pokemon_info = {
            "id": data["id"],
            "name": data["name"],
            "height": data["height"],
            "weight": data["weight"],
            "types": [t["type"]["name"] for t in data["types"]],
            "abilities": [a["ability"]["name"] for a in data["abilities"]],
            "stats": {stat["stat"]["name"]: stat["base_stat"] for stat in data["stats"]},
            "sprites": data["sprites"]
        }
        pokemon_data.append(pokemon_info)

# Verificando os dados coletados
for pokemon in pokemon_data:
    print(pokemon)

# Inserir os dados no MongoDB Atlas
collection.insert_many(pokemon_data)
print("Dados dos Pokémon inseridos com sucesso!")

# Verificar um documento inserido
documento = collection.find_one()
print(documento)

# Exemplo de consultas após inserir os dados
# Consultar por um Pokémon específico pelo nome
resultado = collection.find_one({"name": "pikachu"})
print(resultado)

# Consultar por Pokémon do tipo "fire"
resultados = collection.find({"types": "fire"})
for pokemon in resultados:
    print(pokemon)

# Consultar Pokémon com peso maior que 50
resultados = collection.find({"weight": {"$gt": 50}})
for pokemon in resultados:
    print(pokemon)

# Atualizar a altura de um Pokémon específico
collection.update_one(
    {"name": "pikachu"},
    {"$set": {"height": 0.8}}  # Novo valor de altura
)

# Consultar o documento atualizado e imprimir o resultado
resultado_atualizacao_1 = collection.find_one({"name": "pikachu"})
print("Resultado da Atualização 1:", resultado_atualizacao_1)

# Atualizar a habilidade de todos os Pokémon
collection.update_many(
    {},
    {"$set": {"updated": True}}  # Adicionando um campo 'updated' como exemplo
)

# Consultar e imprimir alguns documentos para verificar a atualização
resultados_atualizacao_2 = collection.find().limit(3)
print("\nResultado da Atualização 2:")
for pokemon in resultados_atualizacao_2:
    print(pokemon)

# Excluir o documento de um Pokémon específico
collection.delete_one({"name": "mew"})

# Consultar para verificar se o Pokémon foi excluído
resultado_exclusao_1 = collection.find_one({"name": "mew"})
print("\nResultado da Exclusão 1: Documento 'Mew' encontrado:", resultado_exclusao_1)

# Excluir todos os Pokémon do tipo "grass"
collection.delete_many({"types": "grass"})

# Consultar para verificar se há Pokémon do tipo "grass"
resultado_exclusao_2 = collection.find({"types": "grass"})
print("\nResultado da Exclusão 2: Pokémon do tipo 'grass':")
for pokemon in resultado_exclusao_2:
    print(pokemon)
```
